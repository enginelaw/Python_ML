### Line Plots

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(1, figsize=(10,5))                   # First figure in the document, 10 units wide and 5 units tall
plt.subplot(1, 2, 1)                            # 1st of 2 graphs in [1row, 2col] format

x = np.linspace(0, 2, 100)                      # Independent Variable
y = 1 + np.sin(2 * np.pi * x)                   # Dependent Functions
z = np.cos(x)

# Plotting both Functions
plt.plot(x, y, linewidth = 3, color = 'blue')
plt.plot(x, z, linewidth = 1, color = 'red')

plt.title('Sinusoidal Graphs')                  # Describing plots
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.legend(['y=1+sin(2*pi*x)', 'z=cos(x)'])

plt.subplot(1, 2, 2)                            # 2nd of 2 graphs in [1row, 2col] format

t = np.arange(0, 2, 0.01)                       # Independent variable
p = t**2                                        # Dependent Functions
q = t**3

# Plotting both Functions
plt.plot(t, p, linestyle = 'solid')
plt.plot(t, q, linestyle = 'dotted')

plt.title('Algebraic Graphs')                   # Describing plots
plt.xlabel('time (s)')
plt.ylabel('height (m)')
plt.legend(['p=t**2', 'q=t**3'])                # legend([list of labels])

plt.show()                                      # showing the figure
```

### Scatter Plots

```python
plt.figure(2)                                   # Second figure in the document

u = np.linspace(0, 2*np.pi, 30)                 # 30 data points in 0 to 2pi radiance angle
v = np.sin(u)                                   # Dependent variable

plt.scatter(u, v, marker = '^')                 # Optional custom custom marker

plt.grid()                                      # Turn on Grid
plt.show()
```

### Bar Graphs

```python
plt.figure(3)                                   # Third figure in the document

m = [1, 4, 7, 10, 13]                           # Data sets
n = [6, 7, 8, 2, 4]

m2 = [2, 5, 8, 11, 14]
n2 = [7, 8, 2, 4, 2]

plt.bar(m, n, color = 'orange', width = 1)
plt.bar(m2, n2, color = 'blue', width = 1)

plt.legend(['Bars 1', 'Bars 2'])
plt.show()
```

### Pie Charts

```python
plt.figure(4)                                   # Fourth figure in the document

names = ['Frogs', 'Hogs', 'Dogs', 'Logs']       # Groups
sizes = [15, 30, 45, 10]                        # Portions for each group totalling to 100

pop_out = [0, 0, 0, 0.2]                        # pop out Logs slice

plt.pie(sizes, labels = names, explode = pop_out, autopct = '%1.0f%%')

plt.show()
```

### 3D Curve Plots

```python
from mpl_toolkits.mplot3d import axes3d
plt.figure(5)
plt.subplot(1,1,1, projection = '3d')

t = np.linspace(0, 10*np.pi, 1000)
st = np.sin(t)
ct = np.cos(t)

plt.plot(st,ct,t, color = 'dodgerblue', linewidth = 3)
plt.show()
```

### 3D Surface and Contour Plots

```python
from matplotlib import cm
fig = plt.figure(6, figsize=(20,8))

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
[X, Y] = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax1 = fig.add_subplot(1,2,1, projection='3d')
ax1.plot_surface(X, Y, Z, cmap = cm.Spectral)

ax2 = fig.add_subplot(1,2,2)
ax2.contour(X, Y, Z, cmap = cm.Spectral) 
plt.show()
```
